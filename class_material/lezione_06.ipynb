## Exercise

Costruiamo una Python classe chiamata 'DataManager' che  caratterizzata da tre metodi:
1. inizializzatore che ha come argomento filepath che ingestiona una stringa;
2. data_reader che eredita dalla classe gli attributi e ritorna il filepath come un pandas dataframe;
3. slicer - metodo statico - che ingestiona il risultato del data_reader e due stringhe, col1 e col2, che indicano le colonne del dataframe sul quale vogliamo fare slicing

```python
class DataManager:
  def __init__(self, filepath):
    self.filepath = filepath

  def data_reader(self):
    df = pd.read_csv(self.filepath)
    return df

  def slicer(self, col1, col2):
    df = self.data_reader()
    return df.loc[:, [col1, col2]]

  @staticmethod
  def slicer(df, col1, col2):
    return df.loc[:, [col1, col2]]

  
  




```

```python
dm = DataManager(
    filepath='https://raw.githubusercontent.com/andreagiussani/Applied_Machine_Learning_with_Python/master/data/election.csv'
    )
```

```python
my_df = dm.data_reader()
```

```python
dm.slicer(my_df, 'state', 'winner').head(2)
```
